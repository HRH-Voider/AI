Question: In the arcade game Galaxian, a player has a 40% chance of hitting an enemy ship with each shot.
If the player fires 5 shots, what is the probability that they hit exactly 3 enemy ships?

Description: Assumes we have a 40% chance of hitting, we are observing the probability of hitting 6 ships.

```python
import ale_py
import gymnasium as gym

# Register the ALE environments
gym.register_envs(ale_py)

# Now check the registry
gym.pprint_registry()
```

```python
import gymnasium as gym
import matplotlib.pyplot as plt
from IPython.display import clear_output, display
import time

def runSim():
    # Create environment
    env = gym.make('ALE/Galaxian-v5', render_mode=None)
    
    # Create figure
    fig, ax = plt.subplots(figsize=(6, 8))
    ax.axis('off')
    
    # Number of simulations to run
    num_simulations = 100
    
    # Run the simulation for `num_simulations` episodes
    observation, info = env.reset()
    total_reward = 0
    hit = 0
    for step in range(100):
        # Take action
        action = env.action_space.sample()
        observation, reward, terminated, truncated, info = env.step(action)
        total_reward += reward
        if reward > 0:
            hit += 1
        
        # Reset if game over
        if terminated or truncated:
            break
    if hit == 3:
        return 1
    else:
        return 0
    
    plt.close()
    env.close()
```

```python
p = 0
for i in range(100):
    print(i, end = "\r", flush = True)
    p += runSim()

p/100
```
