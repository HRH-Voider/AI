Which artists are most frequently found on both top tracks playlist.

For my study, I will be viewing the Spotify top tracks global and top tracks US to find the artists that were shared between the two and the artists that were exclusive to one or the other. I gathered the data by using GPT to turn the 2024 playlists into csv files, and by using class data from the teacher, the teacher used AI to assign genres to their music, which may not accurately reflect what is actually true.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Print current working directory
print(f"Current working directory: {os.getcwd()}")

# List files in the current directory
print("\nFiles in current directory:")
for file in os.listdir():
    print(f"- {file}")

# Now try to load your file with the correct path
try:
    # Try different approaches to locate your file
    file_name = 'Top Tracks Playlists - USA.csv'  # Replace with your actual filename
    
    # Attempt 1: Try direct loading
    print(f"\nAttempting to load: {file_name}")
    if os.path.exists(file_name):
        df = pd.read_csv(file_name)
        print("File loaded successfully!")
        
        # Create the visualization
        sns.set_style("whitegrid")
        plt.figure(figsize=(12, 6))
        ax = sns.countplot(y='AI Genre', data=df, 
                          order=df['AI Genre'].value_counts().index,
                          palette='viridis')
        plt.title('Frequency Distribution of AI Genres', fontsize=16)
        plt.xlabel('Count', fontsize=12)
        plt.ylabel('AI Genre', fontsize=12)
        for i in ax.containers:
            ax.bar_label(i, fmt='%d')
        plt.tight_layout()
        plt.show()
        
        # Display the frequency table
        freq_table = df['AI Genre'].value_counts().reset_index()
        freq_table.columns = ['AI Genre', 'Frequency']
        print(freq_table)
    else:
        print(f"File '{file_name}' not found in current directory.")
        
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# After the existing visualization code, add:

# Create and display artist frequency table
if 'Artist' in df.columns:  # Check if the Artist column exists
    print("\nFrequency of Artists in Playlists:")
    artist_freq_table = df['Artist'].value_counts().reset_index()
    artist_freq_table.columns = ['Artist', 'Frequency']
    
    # Display the top 20 artists (optional, remove limit if you want to see all)
    print(artist_freq_table.head(20))
    
    # Optional: Create a visualization for artist frequency
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Frequency', y='Artist', 
                data=artist_freq_table.head(15),  # Show top 15 artists
                palette='viridis')
    plt.title('Most Frequent Artists in Playlists', fontsize=16)
    plt.xlabel('Count', fontsize=12)
    plt.ylabel('Artist', fontsize=12)
    plt.tight_layout()
    plt.show()
else:
    print("\nArtist column not found in the dataset")
```
