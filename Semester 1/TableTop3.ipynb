What is the probability of drawing 5 monster cards in a row from the deck during the initial draw phase? Assume no other cards have been drawn, the deck has 60 cards and half of them are monster cards.

The sample space I will be observing is any of the possible cards you can pull when drawing five seperate times in yugioh. One trial of the simulation would involve drawing 5 times and checking if each card is a monster card.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Create a deck of 60 cards where the first 30 are monster cards
deck = np.array(['monster' if i < 30 else 'non-monster' for i in range(60)])

# Function to draw 5 cards and check if all are monster cards
def draw_and_check():
    # Shuffle the deck
    shuffled_deck = np.random.permutation(deck)
    # Draw 5 cards
    drawn_cards = shuffled_deck[:5]
    # Check if all 5 are monster cards
    return all(card == 'monster' for card in drawn_cards)

# Run 10000 simulations
results = []
for _ in range(10000):
    results.append(draw_and_check())

# Calculate and display statistics
all_monsters_count = sum(results)
percentage = all_monsters_count / len(results) * 100

print(f"Out of 10000 simulations, all 5 cards were monster cards {all_monsters_count} times ({percentage:.2f}%)")

# Theoretical probability calculation
theoretical_prob = (30/60) * (29/59) * (28/58) * (27/57) * (26/56) * 100
print(f"Theoretical probability: {theoretical_prob:.2f}%")

# Visualize results
plt.figure(figsize=(8, 5))
result_counts = Counter(results)
plt.bar(['Not all monsters', 'All monsters'], 
        [result_counts[False], result_counts[True]])
plt.ylabel('Number of simulations')
plt.title('Results of 10000 simulations')
plt.show()
```

The theoretical probability of drawing five monster cards in a row assuming no other cards have been drawn, and that there is 30 monster cards in a deck of 60, is 2.62% The empirical probability was 2.56%, with all five cards drawn being monster cards 256 times out of 10,000 simulations.
